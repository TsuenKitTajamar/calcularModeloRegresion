Este código realiza lo siguiente:

1. Calcula la desviación típica de altura y peso.
2. Ajusta el modelo de regresión lineal
  Y=a+bX+E.
3. Muestra el coeficiente Re2 para evaluar el ajuste.
4. Grafica los datos y la recta de regresión.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Datos proporcionados (altura en cm, peso en kg)
alturas = np.array([180, 180, 160, 179, 165, 188, 176, 185, 178, 174, 180, 175, 167, 176, 183, 176, 173])
pesos = np.array([80, 73, 50, 56, 63, 92, 73, 83, 80, 80, 80.5, 77, 65, 69, 87, 94, 75])

# Cálculo de la desviación típica (desviación estándar)
desv_tipica_altura = np.std(alturas, ddof=1)
desv_tipica_peso = np.std(pesos, ddof=1)

print(f"Desviación típica de la altura: {desv_tipica_altura:.2f} cm")
print(f"Desviación típica del peso: {desv_tipica_peso:.2f} kg")

# Modelo de regresión lineal
X = alturas.reshape(-1, 1)
Y = pesos
modelo = LinearRegression()
modelo.fit(X, Y)

# Parámetros del modelo
a = modelo.intercept_
b = modelo.coef_[0]
print(f"Modelo de regresión: Y = {a:.2f} + {b:.2f}X")

# Predicciones y evaluación del modelo
Y_pred = modelo.predict(X)
r2 = r2_score(Y, Y_pred)
print(f"Coeficiente de determinación R^2: {r2:.4f}")

# Gráfica de regresión
plt.scatter(alturas, pesos, color='blue', label='Datos reales')
plt.plot(alturas, Y_pred, color='red', label='Recta de regresión')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Regresión Lineal: Peso vs Altura')
plt.legend()
plt.grid(True)
plt.show()
```
